```python
import pandas as pd

# Provided data
data0 = pipe(image=img_path, question="invoice number?")
data1 = pipe(image=img_path, question="Invoice Date?")
data2 = pipe(image=img_path, question="Vendor Name?")
data3 = pipe(image=img_path, question="Vendor address?")
data4 = pipe(image=img_path, question="Customer Name?")
data5 = pipe(image=img_path, question="Customer address?")
data6 = pipe(image=img_path, question="Total Amount?")

# Create a DataFrame with each question and its answer
df = pd.DataFrame({
    'Question': ["invoice number", "Invoice Date", "Vendor Name", "Vendor Address", "Customer Name", "Customer Address", "Total Amount"],
    'Answer': [data0[0]['answer'], data1[0]['answer'], data2[0]['answer'], data3[0]['answer'], data4[0]['answer'], data5[0]['answer'], data6[0]['answer']]
})

# Print the DataFrame
print(df)
```

```python
# Convert the pdfs to images
!apt-get install poppler-utils
!pip install pdf2image Pillow

import os
from pdf2image import convert_from_path
from PIL import Image

def pdf_to_images(input_folder, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get a list of PDF files in the input folder
    pdf_files = [f for f in os.listdir(input_folder) if f.endswith(".pdf")]

    for pdf_file in pdf_files:
        # Construct the full path of the input PDF file
        pdf_path = os.path.join(input_folder, pdf_file)

        # Convert the PDF to a list of PIL images
        images = convert_from_path(pdf_path)

        # Save each page as an image in the output folder
        for i, image in enumerate(images):
            image_path = os.path.join(output_folder, f"{pdf_file}_page_{i+1}.png")
            image.save(image_path, "PNG")

# Example usage:
input_folder = "/content/drive/MyDrive/Training data"  # Replace with the path to your input folder
output_folder = "/content/drive/MyDrive/Training_images_data"  # Replace with the path to your output folder
pdf_to_images(input_folder, output_folder)
```

```python
# Convert the pdfs to images jpg
!apt-get install poppler-utils
!pip install pdf2image Pillow
import os
from pdf2image import convert_from_path
from PIL import Image

def pdf_to_images(input_folder, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get a list of PDF files in the input folder
    pdf_files = [f for f in os.listdir(input_folder) if f.endswith(".pdf")]

    for pdf_file in pdf_files:
        # Construct the full path of the input PDF file
        pdf_path = os.path.join(input_folder, pdf_file)

        # Convert the PDF to a list of PIL images in JPEG format
        images = convert_from_path(pdf_path, fmt='jpeg')

        # Save each page as an image in the output folder
        for i, image in enumerate(images):
            image_path = os.path.join(output_folder, f"{pdf_file}_page_{i+1}.jpeg")
            image.save(image_path, "JPEG")

# Example usage:
input_folder = "/content/drive/MyDrive/Training data"  # Replace with the path to your input folder
output_folder = "/content/drive/MyDrive/Training_images_data_jpg"  # Replace with the path to your output folder
pdf_to_images(input_folder, output_folder)
```
